```python
import pandas as pd
```

```python
file_path = "greendestination (1).csv"
df = pd.read_csv(file_path)

df.head(), df.info()
```

```python
df
```

```python
# Calculate the attrition rate
total_employees = len(df)
employees_left = df[df['Attrition'] == 'Yes'].shape[0]
attrition_rate = (employees_left / total_employees) * 100

print("Attrition Rate of the company is : ")
attrition_rate
```

```python
# Compare mean values of selected columns for employees who left vs. stayed
comparison = df.groupby("Attrition")[["Age", "YearsAtCompany", "MonthlyIncome"]].mean()
comparison
```

```python
# The overall attrition rate is 16.12%.
# Out of 1470 total employees, 237 have left the organization.
```
